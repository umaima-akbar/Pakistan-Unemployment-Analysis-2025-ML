```python
import pandas as pd

# Load dataset
df = pd.read_csv("Pakistan_Poverty_Dataset_2000_2023.csv")

# Show top rows
df.head()
```

```python
df.columns
```

```python
df.info()
df.isnull().sum()
```

```python
df[["Year", "Unemployment Rate (%)"]].describe()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df["Year"], df["Unemployment Rate (%)"], marker='o', color='purple', linewidth=2)

plt.title("Unemployment Rate in Pakistan (2000–2023)", fontsize=14)
plt.xlabel("Year")
plt.ylabel("Unemployment Rate (%)")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
